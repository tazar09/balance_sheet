```python
df = pd.read_excel('/content/bs1.xlsx')
df = df.dropna(axis = 0, how = 'all')
# df = df.dropna(axis = 0, thresh = 9)
df = df.dropna(axis = 1, how = 'all')
df = df.fillna('')
```

```python
columns = []
for i in range(13):
  columns.append('column' + str(i))
df.columns = columns
```

```python
rows_indexes = list(df[df['column7'] == 'Pagina'].index)
df = df.drop(axis = 0, labels = rows_indexes)
```

```python
df.loc[0,'column0'] = df.loc[0,'column5'][:-2].title()
df.loc[2,'column0'] = df.loc[2,'column0'].title()
df.loc[0,'column5'] = ''
df.loc[0,'column1'] = str(df.loc[0,'column8']) + str(df.loc[0,'column10'])
df.loc[0,'column8'] = ''
df.loc[0,'column10'] = ''
df.loc[2, 'column1'] = str(int(df.loc[2, 'column1']))
df.loc[4, 'column1'] = df.loc[4, 'column0'].split(':')[1].strip()
df.loc[4, 'column0'] = df.loc[4, 'column0'].split(':')[0].strip()
```

```python
def formular(value):
  if 'FORMULAR' in value:
    return value.title()
  else:
    return value
```

```python
df['column0'] = df['column0'].apply(lambda x: formular(x))
```
